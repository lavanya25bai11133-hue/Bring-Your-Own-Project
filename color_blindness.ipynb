```python
def simulate_color_blindness(image, blindness_type):

    protanopia_matrix = np.array([
        [0.567, 0.433, 0.000],
        [0.558, 0.442, 0.000],
        [0.000, 0.242, 0.758]
    ])


    deuteranopia_matrix = np.array([
        [0.625, 0.375, 0.000],
        [0.700, 0.300, 0.000],
        [0.000, 0.300, 0.700]
    ])


    tritanopia_matrix = np.array([
        [0.950, 0.050, 0.000],
        [0.000, 0.433, 0.567],
        [0.000, 0.475, 0.525]
    ])

    img_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB).astype(np.float32) / 255.0


    if blindness_type == 'protanopia':
        transformed_img = np.dot(img_rgb.reshape(-1, 3), protanopia_matrix.T).reshape(img_rgb.shape)
    elif blindness_type == 'deuteranopia':
        transformed_img = np.dot(img_rgb.reshape(-1, 3), deuteranopia_matrix.T).reshape(img_rgb.shape)
    elif blindness_type == 'tritanopia':
        transformed_img = np.dot(img_rgb.reshape(-1, 3), tritanopia_matrix.T).reshape(img_rgb.shape)
    else:
        raise ValueError("Invalid blindness_type. Choose from 'protanopia', 'deuteranopia', 'tritanopia'.")


    transformed_img = np.clip(transformed_img, 0, 1) * 255
    return cv2.cvtColor(transformed_img.astype(np.uint8), cv2.COLOR_RGB2BGR)
```

```python
import gradio as gr
from PIL import Image
import numpy as np
import cv2

def gradio_simulate_color_blindness(input_image, blindness_type):

    img_np = np.array(input_image)
    img_cv = cv2.cvtColor(img_np, cv2.COLOR_RGB2BGR)

    simulated_image_cv = simulate_color_blindness(img_cv, blindness_type)

    simulated_image_rgb = cv2.cvtColor(simulated_image_cv, cv2.COLOR_BGR2RGB)
    return Image.fromarray(simulated_image_rgb)



interface = gr.Interface(
    fn=gradio_simulate_color_blindness,
    inputs=[
        gr.Image(type="pil", label="Upload Image"),
        gr.Dropdown(["protanopia", "deuteranopia", "tritanopia"], label="Select Blindness Type", value="protanopia")
    ],
    outputs=gr.Image(type="pil", label="Simulated Color Blindness"),
    title="Color Blindness Simulator",
    description="Upload an image and select a type of color blindness to see the simulated result."
)

interface.launch(debug=True, share=True)
```
